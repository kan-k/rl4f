# Optimal Trade Execution with Reinforcement Learning

## Educational Overview

This notebook tackles one of the most practical problems in quantitative trading: **optimal execution**. When you need to buy or sell a large quantity of shares, how do you minimize the total cost while managing market impact?

### The Optimal Execution Problem:

**Real-World Challenge:**
- **Large orders** cannot be executed instantly without moving prices
- **Market impact**: Your own trading affects the stock price
- **Trade-off**: Speed vs. cost vs. risk
- **Timing decisions**: When and how much to trade at each step

**Traditional Approach - Almgren-Chriss Model:**
- **Mathematical optimization** with closed-form solutions
- **Linear market impact** assumptions
- **Fixed parameters** throughout execution period
- **Risk-averse** objective function balancing cost and risk

### Key Components of Execution Costs:

**1. Temporary Market Impact:**
- **Immediate price movement** due to trade size
- **Bid-ask spread** effects
- **Liquidity consumption** costs

**2. Permanent Market Impact:**
- **Persistent price changes** from your trading
- **Information leakage** - market learns about your intentions
- **Cumulative effect** of all your trades

**3. Execution Risk:**
- **Price uncertainty** during the execution period
- **Volatility risk** - prices may move against you
- **Timing risk** - when to execute remaining shares

### Why Use Reinforcement Learning?

**Advantages over Traditional Methods:**
- **Adaptive strategies**: Learns from changing market conditions
- **Non-linear relationships**: Captures complex market dynamics
- **Real-time adaptation**: Adjusts strategy based on market feedback
- **No parameter assumptions**: Learns optimal behavior through experience

2. **Cost decomposition** - understanding execution costs
3. **RL environment design** for execution problems
4. **Agent training** to minimize total execution costs

## The Almgren-Chriss Model: Theoretical Foundation

The **Almgren-Chriss model** (2001) provides the mathematical foundation for optimal trade execution. It balances **execution costs** against **market risk** to find the optimal trading strategy.

### Mathematical Framework:

**Objective Function:**
The model minimizes: **E[Cost] + λ × Var[Cost]**

Where:
- **E[Cost]**: Expected execution cost
- **Var[Cost]**: Variance of execution cost (risk)
- **λ**: Risk aversion parameter (λ ↑ = more risk-averse)

### Key Parameters:

**Market Parameters:**
- **T**: Total execution time horizon
- **N**: Number of trading periods
- **σ**: Stock price volatility
- **S₀**: Initial stock price

**Trading Parameters:**
- **X**: Total shares to execute
- **γ**: Permanent market impact coefficient
- **η**: Temporary market impact coefficient
- **λ**: Risk aversion parameter

### Cost Components:

**1. Temporary Impact (η):**
- **Immediate cost** of each trade
- **Proportional to trade rate** (shares/time)
- **Recovers after trade** (temporary price displacement)

**2. Permanent Impact (γ):**
- **Persistent price movement** from trading
- **Cumulative effect** of all trades
- **Information content** of your order flow

**3. Market Risk (λσ²):**
- **Price volatility** during execution period
- **Remaining inventory risk** - unexecuted shares face price risk
- **Trade-off**: Execute quickly (reduce risk) vs. slowly (reduce impact)

### Optimal Solution:
The model provides a **closed-form solution** showing how to optimally split your order across time to minimize total expected cost plus risk penalty.

This theoretical benchmark allows us to evaluate how well our RL agent learns optimal execution strategies.

```python
class AlmgrenChriss:
    def __init__(self, T, N, S0, sigma, X, gamma, eta, lamb):
        self.T = T              
        self.N = N           
        self.dt = T / N
        self.S0 = S0
        self.sigma = sigma
        self.X = X
        self.gamma = gamma
        self.eta = eta
        self.lamb = lamb
```

```python
class AlmgrenChriss(AlmgrenChriss):
    def optimal_execution(self):
        kappa = np.sqrt(self.lamb * self.sigma ** 2 / self.eta)
        t = np.linspace(0, self.T, self.N + 1)
        xt_sum = (self.X * np.sinh(kappa * (self.T - t)) /
                  np.sinh(kappa * self.T))
        xt = -np.diff(xt_sum, prepend=0)
        xt[0] = 0
        return t, xt
```

```python
class AlmgrenChriss(AlmgrenChriss):
    def simulate_stock_price(self, xt, seed=None):
        rng = default_rng(seed=seed)
        S = np.zeros(self.N + 1)
        S[0] = self.S0
        P = np.zeros(self.N + 1)
        P[0] = self.S0
        for t in range(1, self.N + 1):
            dZ = rng.normal(0, np.sqrt(self.dt))
            S[t] = S[t - 1] + sigma * dZ
            P[t] = S[t] - self.gamma * xt[:t + 1].sum()
        return S, P
```

```python
class AlmgrenChriss(AlmgrenChriss):
    def calculate_costs(self, xt):
        temporary_cost = np.sum(self.eta *
                    (xt / self.dt) ** 2 * self.dt)
        permanent_cost = np.sum(self.gamma * np.cumsum(xt) * xt)
        execution_risk = self.lamb * self.sigma ** 2 * np.sum(
            (np.cumsum(xt[::-1])[::-1] / self.dt) ** 2 * self.dt)
        TEC = temporary_cost + permanent_cost + execution_risk
        return temporary_cost, permanent_cost, execution_risk, TEC
```

## Understanding Execution Costs and Trade-offs

Before building our RL environment, let's examine how different risk aversion parameters (λ) affect optimal execution strategies and costs.

### Key Insights from Almgren-Chriss Results:

**High λ (Risk-Averse Strategy):**
- **Front-loaded execution**: Trade more early, less later
- **Reduces inventory risk**: Less exposure to price movements
- **Higher market impact**: Aggressive trading increases costs
- **Lower execution risk**: Minimizes variance in final cost

**Low λ (Cost-Focused Strategy):**
- **Uniform execution**: Spread trades more evenly over time
- **Minimizes market impact**: Gentle trading reduces price displacement
- **Higher inventory risk**: More exposure to adverse price movements
- **Lower expected cost**: But higher variance in outcomes

### The Fundamental Trade-off:

**Speed vs. Cost vs. Risk**
- **Execute quickly**: Low risk, high impact cost
- **Execute slowly**: Low impact cost, high market risk
- **Optimal strategy**: Balances these competing objectives

### Why This Matters for RL:

**1. Benchmark Comparison:**
- RL strategies should approach or beat theoretical optimal
- Provides performance evaluation criteria

**2. Market Reality:**
- Real markets have **non-linear impact** functions
- **Parameter instability** - market conditions change
- **Liquidity variations** throughout the day
- **Information asymmetries** not captured in linear models

**3. RL Advantages:**
- **Adaptive execution**: Learns from market feedback
- **Non-parametric**: No assumptions about impact functions
- **Context-aware**: Adjusts to changing market conditions

The following environment enables our RL agent to learn execution strategies that potentially improve upon traditional approaches.

```python
class Execution:
    def __init__(self, T, N, sigma, X, gamma, eta, lamb):
        self.T = T              
        self.N = N           
        self.dt = T / N
        self.sigma = sigma
        self.X = X
        self.gamma = gamma
        self.eta = eta
        self.lamb = lamb
        self.episode = 0
```

```python
class Execution(Execution):
    def _get_state(self):
        s = np.array([self.X_,
                    self.bar / self.N])
        state = np.hstack((self.xt, s))
        return state, {}
    def reset(self):
        self.bar = 0
        self.treward = 0
        self.episode += 1
        self.X_ = self.X
        self.xt = np.zeros(self.N + 1)
        self.tec = pd.DataFrame(
            {'pc': 0, 'tc': 0, 'er': 0}, index=[0])
        return self._get_state()
```

```python
class Execution(Execution):
    def step(self, action):
        self.bar += 1
        self.xt[self.bar] = action
        self.X_ -= action
        pc = np.sum(self.gamma *
                np.cumsum(self.xt) * self.xt)
        tc = np.sum(self.eta *
                (self.xt / self.dt) ** 2 * self.dt)
        er = self.lamb * self.sigma ** 2 * np.sum(
            (np.cumsum(self.xt[::-1])[::-1] / self.dt) ** 2
            * self.dt)
        df = pd.DataFrame({'pc': tc, 'tc': pc, 'er': er},
                          index=[0])
        self.tec = pd.concat((self.tec, df))
        cost = self.tec.diff().fillna(0).iloc[-1]
        tec = cost.sum()
        self.state, _ = self._get_state()
        pen = 0
        if self.bar < self.N:
            if self.X_ <= 0.0001:
                done = True
            else:
                done = False
        elif self.bar == self.N:
            pen = abs(self.X_) * 10
            done = True
        return self.state, -(tec + pen), done, False, {}
```

## Designing an RL Environment for Optimal Execution

Now we create a custom **reinforcement learning environment** that allows an agent to learn optimal execution strategies through trial and error.

### Environment Design Philosophy:

**State Space**: What does our agent observe?
- **xt vector**: Trade sizes executed at each time step
- **X_**: Remaining shares to execute
- **t/T**: Fraction of time period completed
- **Cumulative execution history**: Learn from past decisions

**Action Space**: What decisions can our agent make?
- **Trade size**: How many shares to execute this period
- **Continuous actions**: Real numbers (not just discrete choices)
- **Constraints**: Cannot exceed remaining inventory

**Reward Function**: How do we measure success?
- **Negative total execution cost**: -(Temporary + Permanent + Risk costs)
- **Penalty for incomplete execution**: Heavy penalty if shares remain at end
- **Real-time feedback**: Immediate cost assessment after each trade

**Episode Structure**:
- Each episode = one complete execution sequence
- Agent learns from thousands of execution experiences
- Random initialization to improve generalization

### Key RL Challenges in Execution:

1. **Sequential decision making**: Each trade affects future options
2. **Constraint handling**: Must execute all shares by deadline
3. **Cost minimization**: Lower rewards = better performance
4. **Path dependency**: Order of trades matters for total cost

### Advanced Agent Features:

**Actor-Critic Architecture:**
- **Actor network**: Learns execution policy (trade sizes)
- **Critic network**: Evaluates state values
- **Simultaneous learning**: Both networks improve together

**Smart Initialization:**
- **Dirichlet distribution**: Ensures valid execution sequences
- **Different priors**: Equal-weight vs. front-loaded vs. back-loaded
- **Exploration strategies**: Balanced learning across approaches

This environment enables our agent to discover execution strategies that potentially outperform traditional mathematical optimization.

```python
class ExecutionAgent(DQLAgent):
    def __init__(self, symbol, feature, n_features, env,
                 hu=24, lr=0.0001, rng='equal'):
        self.epsilon = 1.0
        self.epsilon_decay = 0.9975
        self.epsilon_min = 0.1
        self.memory = deque(maxlen=2000)
        self.batch_size = 32
        self.eta = 1.0
        self.trewards = list()
        self.max_treward = -np.inf
        self.n_features = n_features
        self.env = env
        self.episodes = 0
        self.rng = rng
        self._generate_rn()
        self.actor = self._create_model(hu, lr, 'sigmoid')
        self.critic = self._create_model(hu, lr, 'linear')
```

```python
class ExecutionAgent(ExecutionAgent):
    def _generate_rn(self):
        if self.rng == 'equal':
            alpha = np.ones(self.env.N)
        elif self.rng == 'decreasing':
            alpha = range(self.env.N, 0, -1)
        else:
            alpha = rng.random(self.env.N)
        rn = rng.dirichlet(alpha)
        self.rn = np.insert(rn, 0, 0)
```

```python
class ExecutionAgent(ExecutionAgent): 
    def _create_model(self, hu, lr, out_activation):
        model = Sequential()
        model.add(Dense(hu, input_dim=self.n_features,
                        activation='relu'))
        model.add(Dense(hu, activation='relu'))
        model.add(Dense(1, activation=out_activation))
        model.compile(loss='mse', optimizer=opt(learning_rate=lr))
        return model
```

```python
class ExecutionAgent(ExecutionAgent):    
    def act(self, state):
        if random.random() <= self.epsilon or self.episodes < 250:
            return min(self.rn[self.f], state[0, -2])
        else:
            action = self.actor.predict(state)[0, 0]
        return action
```

```python
class ExecutionAgent(ExecutionAgent):
    def replay(self):
        batch = random.sample(self.memory, self.batch_size)
        for state, action, next_state, reward, done in batch:
            target = reward
            if not done:
                target += self.eta * self.critic.predict(
                        next_state)[0, 0]
                self.critic.fit(state, np.array([target]),
                        epochs=1, verbose=False)
                # advantage = target - self.critic.predict(state)[0, 0]
                self.actor.fit(state, np.array([action]),
                        # sample_weight=np.array([advantage]),
                        epochs=1, verbose=False) 
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay
        self._generate_rn()
```

```python
class ExecutionAgent(ExecutionAgent):
    def test(self, episodes, verbose=True):
        for e in range(1, episodes + 1):
            state, _ = self.env.reset()
            state = self._reshape(state)
            treward = 0
            for _ in range(1, self.env.N + 1):
                action = self.actor.predict(state)[0, 0]
                state, reward, done, trunc, _ = self.env.step(action)
                state = self._reshape(state)
                treward += reward
                if done:
                    templ = f'total reward={treward:4.3f}'
                    if verbose:
                        print(templ)
                    break
            print(self.env.xt)
```
